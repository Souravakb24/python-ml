```python
from matplotlib import pyplot as plt
```

line plot

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7,
14958.3]

plt.plot(years, gdp, color='b', marker='o', linestyle='solid')
plt.title("National GDP")
plt.xlabel("Year")
plt.ylabel("Billions of $")
plt.show()  # to show the plot
# plt.savefig()  to save the plot
```

Bar chart

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi",
"West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(movies, num_oscars, color='green')
plt.title("My favourite Movies")
plt.xlabel("Movie")
plt.ylabel("# of Oscars")
plt.show()
```

```python
plt.bar(range(len(movies)), num_oscars, color='green')
plt.title("My favourite Movies")
plt.ylabel("# of Oscars")

plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
from collections import Counter

grades=[83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
'''Bucket grades by decile, but put 100 in with the 90s'''
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()], histogram.values(), 10,  edgecolor=(0, 0, 0), color='orange')
# (shift 5 for every key to maintain in range)              (width of bar) (bar edge colour = black)
plt.axis([-5, 105, 0, 5]) #x-axis from -5 to 105
plt.xticks([10 * i for i in range(11)])  #x-axis labels at 0, 10,...,100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

Line chart

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x+y for x,y in zip(variance,bias_squared)]
print("total_error ",total_error)
xs = [i for i,_ in enumerate(variance)]
print("xs ",xs)

plt.plot(xs,variance , 'b-', label='variance')
plt.plot(xs, bias_squared , 'r-.', label='bias^2')
plt.plot(xs,total_error, 'g:', label='total error')

plt.legend(loc=9)
```

Scatter plot

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a','b','c','d','e','f','g','h','i']
plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels,friends, minutes):
    plt.annotate(label,xy=(friend_count, minute_count),xytext=(5, -5),textcoords='offset points')
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
